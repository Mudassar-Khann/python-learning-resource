# Loops in Python

Loops help you repeat code efficiently. This notebook covers core loop concepts and practical patterns used in real programs.

Topics:
- `for` loop
- `while` loop
- `break`, `continue`, `pass`
- nested loops
- extra important loop tools (`for-else`, `enumerate`)


## 1. `for` Loop (Iteration over sequence)

Used when the number of iterations is known, or when iterating over an existing sequence.


```python
for i in range(5):
    print(i)
```

Output:

`0 1 2 3 4`

### `range(start, stop, step)`

- `start` defaults to `0`
- `step` defaults to `1`
- `stop` is excluded

Example: `range(2, 10, 2)` gives `2, 4, 6, 8`.


```python
print(list(range(2, 10, 2)))
print(list(range(10, 0, -2)))

names = ["Ava", "Noah", "Liam"]
for name in names:
    print(name)
```

## 2. `while` Loop (Condition-controlled)

Used when number of iterations is unknown.


```python
x = 0
while x < 5:
    print(x)
    x += 1
```

Important: update the loop variable, otherwise the loop can run forever (infinite loop).

`while True` is useful when you want to keep asking for input until a condition is met.


```python
while True:
    x = input("Enter q to quit: ")
    if x == "q":
        break
```

## 3. `break`

Stops the loop immediately.


```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

Output: `0 1 2 3 4`

Real use: stop searching once the target is found.


```python
numbers = [4, 9, 12, 20, 31]
target = 12

for n in numbers:
    if n == target:
        print("Found", target)
        break
```

## 4. `continue`

Skips the current iteration and moves to the next one.


```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

Output: `0 1 3 4`


## 5. `pass`

Does nothing. It is a placeholder where Python expects a statement.


```python
for i in range(3):
    pass

def todo_feature():
    pass
```

## 6. Nested Loops
A loop inside another loop.

A nested loop is a loop inside another loop, where the inner loop runs completely for every iteration of the outer loop.

#### Visual intuition
Think of a clock.

hour loop:
        
        minute loop:

For every hour, the minute loop runs fully.


```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

```python
for i in range(5):
    print(i, end=" " )
    for j in range(10,12):
        print( j)
```

```python
#matric traversal
matrix = [
    [1,2,3],
    [4,5,6]
]

for row in matrix:
    for value in row:
        print(value , end=" ")
    print()
```

## 7. Looping Through Dictionaries

A dictionary stores key-value pairs.

if you don't know how dictionary works first study it through **06_datastructures.ipynb**

```python
user = {
    "name" : "thomas ",
    "age" : 18,
    "city" : "paris"
}
for key in user:
    print(key , user[key])
```

```python
# using key method of dict
user = {
    "name" : "thomas ",
    "age" : 18,
    "city" : "paris"
}
for key in user.keys():
    print(key )
```

```python
# using values method of dict
user = {
    "name" : "thomas ",
    "age" : 18,
    "city" : "paris"
}
for key in user.values():
    print(key )
```

```python
# using item method of dict which print both key and value
user = {
    "name" : "thomas ",
    "age" : 18,
    "city" : "paris"
}
for key , value in user.items():
    print(key, value )
```

## Extra Important Concepts

### `for-else`

`else` runs only if loop ends normally (no `break`). Useful in search/check tasks.


```python
target = 2
numbers = [2, 4, 6, 8]

for n in numbers:
    if n == target:
        print("Found")
        break
else:
    print("Not found")
```

```python
for x in range(6):
  if 4 == x:
    break
  print(x)
else:
  print("Finally finished!")
```

### `enumerate()`

Use `enumerate` when you need both index and value.

enumerate() lets you loop over a sequence while automatically getting the index of each item.

Think of enumerate() as attaching numbers to items.

it return tuples where it attaches index to value. 

with start we can tell from where to start counter. 

```python
fruits = ["apple", "banana", "mango"]
for index, fruit in enumerate(fruits, start= 1):
    print(index, fruit)
```

#### Comparison 

```python
# Without enumerate
names = ["Ali","Sara","John"]
for i in range(len(names)):
    print(i , names[i])
```

```python
# With enumerate
names = ["Ali","Sara","John"]
for index , value in enumerate(names, start= 1):
    print(index , value )
```

### `zip()`
One-sentence idea

zip() lets you loop over multiple sequences simultaneously.

```python
names = ["Ali","Sara","John"]
scores = [90,85,88]

for name , score in zip(names, scores):
    print(name, score)
```

## Summary

- Use `for` for known/repeatable sequences.
- Use `while` for condition-based repetition.
- Use `break` to stop early, `continue` to skip, `pass` as placeholder.
- Use nested loops for grid/pattern logic.
- Use `for-else` and `enumerate` and `zip` for cleaner real-world code.
