# Loops

This notebook covers looping structures:
- for loops
- while loops
- break and continue statements

### for Loop (Iteration over sequence)

Used when number of iterations is known

```python
for i in range(5):
    print(i, end=' ')

print()
for i in range(5, 20, 2):
    print(i)
```

**range(start, stop, step)**

### while Loop (Condition-controlled)

Used when iterations unknown

```python
x = 0
while x < 5:
    print(x)
    x += 1
    # Must change variable or → infinite loop
```

```python
# break - Stops loop immediately

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue - Skips current iteration

for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# pass - Does nothing (placeholder)

for i in range(3):
    pass
```

```python
# Nested Loops - Loop inside another loop

for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# Sum numbers
total = 0
for i in range(1, 6):
    total += i
print(total)
```

```python
# Input loop
while True:
    x = input("Enter 'q' to quit: ")
    if x == "q":
        break
```

```python
# Skip even numbers
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```
