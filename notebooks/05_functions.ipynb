# Functions

A function is a named, reusable block of code that:

- Can receive input (parameters)
- Can process data
- Can return output
- Helps organize logic
- Prevents repetition


## 1. Why Functions Exist

Without functions, logic gets repeated and code becomes harder to manage.


```python
# Repeated logic (without a function)
a = 5
print(a * a)

b = 10
print(b * b)

# Reusable logic (with a function)
def square(n):
    return n * n

print(square(5))
print(square(10))
```

Functions improve:

- Reusability
- Readability
- Maintainability
- Testing


## 2. Basic Syntax


```python
def function_name(parameters):
    # body
    return value


def greet(name):
    return f"Hello {name}"

print(greet("Ali"))
# As we know it returns a string so we can apply strings methods to what it ruturns
print(greet("Ali").upper())
```

## 3. Function Anatomy


```python
def add(a, b):
    result = a + b
    return result
```

| Part | Meaning |
| --- | --- |
| `def` | defines a function |
| `add` | function name |
| `a, b` | parameters |
| body | logic inside function |
| `return` | sends value back |


## 4. Parameters vs Arguments

Parameters are placeholders inside the function definition.
Arguments are actual values passed when calling the function.


```python
def multiply(x, y):  # parameters
    return x * y

print(multiply(3, 4))  # arguments
```

## 5. Return vs Print (Critical Difference)

`print()` only displays a value. `return` gives a value back to the caller.


```python
def f_print():
    print(5)

x = f_print()
print(x)  # None, because f_print does not return a value
```

```python
def f_return():
    return 5
# A return gives the value can be stored in a variable and can be modified or used later and return also exit the function

x = f_return()
print(x)
```

## 6. What If No Return?

Every function returns something. If you do not write `return`, Python returns `None`.


```python
def f_no_return():
    pass

print(f_no_return())
```

## 7. Multiple Return Values

A function can return multiple values, which are returned as a tuple.


```python
def calc(a, b):
    return a + b, a - b

print(calc(8, 3))
```

```python
# multiple returend values can also be stored in multiple veriables
# now they belongs to that veriable insted of being in tuple

def calc(a, b):
    return a + b, a - b
add , min = calc(8, 3)
print(add, min)
```

## 8. Default Parameters

if no argument passed function moves to deafult value that we have set in function parameters


```python
def greet_default(name="Guest"):
    return f"Hello {name}"

print(greet_default())
print(greet_default("Sara"))
```

```python
# example with numbers
def add(a=2, b=3):
    return a + b

print(add())
print(add(5,5))
```

## 9. Keyword Arguments

With keyword arguments, order does not matter because names are explicit.


```python
def divide(a, b):
    return a / b

print(divide(b=2, a=10))
```

## 10. Scope (Intro)

Variables created inside a function are local to that function.


```python
x = 10

def show_scope():
    x = 5
    print(x)

show_scope()
print(x)
```

## 11. Functions Are First-Class Objects

Functions can be assigned to variables, passed around, and returned.

(`A first-class object is something in a programming language that can be treated like normal data: stored in variables, passed to functions, and returned from functions.`)


```python
def greet_short():
    return "hi"

x = greet_short
print(x())
```

#### Important detail:

x = greet_short

means

`"store the function itself"`

NOT

x = greet_short()

which means

`"run the function and store the result"`

##### `Think of it like a machine vs using the machine.`

blender      → the machine

blender()    → turning the blender on

```python
# another example
def add(a, b):
    return a + b
a = add
print(a(2,3))
```

#### Some Examples

```python
def add(a, b):
    return a + b
def run_operation(func):
    return func(3,4)

run_operation(add)
```

**WHY WE PASS FUNCTIONS WITHOUT PARENTHESES**

Because if we add parentheses we run the function immediately.

But sometimes we want to give the function to another function so it can run it later.

```python
def add(a, b):
    return a + b
def run_operation(func):
    x = func
    return x(2,4)

run_operation(add)
```

#### Building Flexible Systems

Example: operation selector.

```python
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

def calculate(func, a, b):
    return func(a, b)

print(calculate(add, 2, 3))
print(calculate(multiply, 2, 3))
```

```python
num = [-5, -3, 1, -4, 7,-8]

sorted_numbers = sorted(num, key=abs)

print(sorted_numbers)
```

### Scope Global and Local Variables

Variables defined outside functions are global veriables while veriable defined inside scope of function 
are called Local veriable.

```python
x = 10
s = 30
g = 90
def show():
    g = 0
    m = 80
    print(x, s, m, g)

show()
# Function can read global variables but cannot modify it
```

#### Modifying Global Variables

```python
x = 10

def change():
    x = 20

    # This fails Because Python creates a new local variable.
```

```python
# To modify the global variable:

x = 10
s = 90
n = 50
def change():
    global x , s , n
    x = 20
    s = 1
    n = 2
    print(x, s, n)

change()
print(x, s, n)
```

## 12. Common Beginner Mistakes

- Forgetting `return`
- Using `print` instead of `return`
- Modifying globals unintentionally


```python
def add_wrong(a, b):
    a + b  # Missing return, so this function returns None

print(add_wrong(2, 3))
```

## 13. Real Example - Clean Program

Keep logic inside the function and decisions outside.


```python
def is_even(n):
    return n % 2 == 0

num = int(input("Enter number: "))

if is_even(num):
    print("Even")
else:
    print("Odd")
```
