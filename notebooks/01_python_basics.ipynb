# Python Basics

This notebook covers fundamental Python concepts:
- Arithmetic Operators
- Assignment Operators  
- Comparison Operators

**Operator Precedence (important)**

2 + 3 * 4 = 14
(2 + 3) * 4 = 20

### Comparison Operators

| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | equal            |
| `!=`     | not equal        |
| `>`      | greater          |
| `<`      | less             |
| `>=`     | greater or equal |
| `<=`     | less or equal    |

```python
x = 5
x = x + 5
print(x)
x = 5
# is same as 

x += 5
print(x)
```

### Assignment Operators

| Operator | Meaning               |
| -------- | --------------------- |
| `=`      | assign                |
| `+=`     | add & assign          |
| `-=`     | subtract & assign     |
| `*=`     | multiply & assign     |
| `/=`     | divide & assign       |
| `//=`    | floor divide & assign |
| `%=`     | remainder & assign    |
| `**=`    | power & assign        |

### Arithmetic Operators

| Operator | Meaning             | Example | Result |
| -------- | ------------------- | ------- | ------ |
| `+`      | Add                 | `5+2`   | `7`    |
| `-`      | Subtract            | `5-2`   | `3`    |
| `*`      | Multiply            | `5*2`   | `10`   |
| `/`      | Division (float)    | `5/2`   | `2.5`  |
| `//`     | Floor division      | `5//2`  | `2`    |
| `%`      | Modulus (remainder) | `5%2`   | `1`    |
| `**`     | Power               | `2**3`  | `8`    |
