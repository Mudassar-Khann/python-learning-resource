### Sets

A set is:
- Unordered
- Mutable
- Unique elements only
- Fast membership testing
- Implemented using hash table

```python
a = {1,2,3}
b = {3,4,5}
```

Set Operations

Let a = {1,2,3} and b = {3,4,5}

```python
# Creating Sets
s1 = {1,2,3}
s2 = set([1,2,3])
s3 = set()     # empty set

# {} is empty dictionary, not set

# Uniqueness
s = {1,1,2,2,3}
# Result: {1,2,3}

# Membership
3 in s

# Add / Remove
s.add(10)

s.remove(2)    # error if not found
s.discard(5)   # safe, no error

a = {1,2,3}
b = {3,4,5}
```

### Dictionaries

A dictionary is:
- Unordered (in concept; insertion-ordered in modern Python)
- Mutable
- Key â†’ Value mapping
- Keys must be immutable & unique
- Fast lookup (hash table)

```python
# Nested Dictionary
student = {
    "name": "Ali",
    "marks": {
        "math": 90,
        "cs": 95
    }
}

print(student["marks"]["math"])
```

```python
# Iteration
d = {"snake" : "animal", "eagle" : "bird"}
for k in d:
    print(k, d[k])

for k,v in d.items():
    print(k, v)
```

```python
# Creating Dictionaries
d1 = {}
d2 = dict()
d3 = {"a":1, "b":2}

# Accessing Values
d = {"a":10, "b":20}

d["a"]        # 10
d.get("a")    # 10
d.get("x")    # None (safe)

# Direct indexing = KeyError if missing
s = {}

# Creating/Updating
s["age"] = 25      # add
s["age"] = 30      # update

# Deleting 
del s["age"]
s["max"] = 19
s.pop("max")
```

```python
d = {
    "name": "Ali",
    "age": 21,
    "is_student": True
}
```

### Lists

A list is:
- Ordered
- Mutable
- Allows duplicates
- Heterogeneous (mixed types allowed)

```python
# Tuple Unpacking
t = (10,20,30)
a, b, c = t
print(a, b, c)

# Partial unpacking
a, *b = (1,2,3,4)
print(a)
print(b)
```

```python
# Indexing
t = (10,20,30)
t[0]   #10
t[-1]   #30

# Immutability
t = (1,2,3)
t[0] = 5   # ERROR

# But mutable objects inside tuple CAN change:
t = ([1,2], 3)
t[0].append(5)   # allowed

# Tuple is immutable, not its contents
```

```python
t = (5)
s = (5,)
print(type(t))
print(type(s))
```

```python
# Creating Tuples
a = (1,2,3)
b = tuple([1,2,3])

# Single-element tuple (important)
t = (5,)   # comma required

# Without comma it is considered integer, not tuple
```

```python
# Creating Lists
a = []
b = list()
c = [1,2,3,4]

# Indexing
a = [10,20,30,40]
a[0]   #10
a[-1]  #40

# Slicing
a = [0,1,2,3,4,5]

a[1:4]   
a[:3]    
a[::2]   
a[::-1]
```
