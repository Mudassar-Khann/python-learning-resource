# Conditionals and Booleans

This notebook covers:
- Conditionals (if-elif-else)
- Boolean Logic
- Operators (logical, identity, membership)

## Conditionals

Conditionals control decision making.

Basic Syntax:
```
if condition (runs if condition True ):
    block
elif condition (runs if if block is false and this one is true ):
    block
else (runs if both blocks are false there is no condition for else it runs if previous blocks are false ):
    block
```

if → runs block only when condition is True

if-else → one of two blocks runs

if-elif-else → checks multiple conditions in order, first True block runs

````python
x = 0
if x:
    print("A")
else:
    print("B")
````

Reason: 0 is falsy boolean expression, that's why else block runs

### Truthiness Reminder

False conditions include:
- 0, 0.0
- ""
- None
- Empty containers e.g {}, [], ""

Everything else → True

````python
# == (Equal to) checks value equality, not memory

5 == 5        # True
"hi" == "hi"  # True
5 == 5.0      # True (values equal)
````

````python
name = "mudassar"
# double equal is boolean logic, it either evaluates to true or false 
if name == "mudassar":
    print("ok it runs")
````

````python
# Boolean logic 
user_answer = input("have you logged into your account? yes or no: ")

if user_answer == "yes":
    print("give access of account to user")
elif user_answer == "no":
    print("secure user account")
else:
    print("please enter correct message")
````

````python
# Using not
x = False

if not x:
    print("hello")
````

### Logical Operators

| Operator | Meaning           |
| -------- | ----------------- |
| `and`    | both must be True |
| `or`     | at least one True |
| `not`    | reverse           |

### Identity Operator — is vs ==

- `is` → checks same object in memory
- `==` → checks same value

````python
a = [1,2]
b = [1,2]

a == b   # True (same value)
a is b   # False (different objects)

s = 10
n = 10

s is n 

if s is n:
    print("working")
````

When **is** SHOULD be used:
- Only for: None
- Singleton objects (True, False)

````python
x = None

if x is None:
    print("No value")
````

````python
# Avoid:
if x == None:   # not recommended
    pass
````

### Membership Operator — in

Checks containment in sequences

````python
"a" in "apple"        # True
3 in [1,2,3,4]        # True
"x" not in "apple"    # True
````

Used in:
- Searching
- Filtering
- Validation
- Loops
- Conditionals

Example:

````python
password = "admin123"

if "123" in password:
    print("Weak password")
````

````python
# Chained Comparisons 
x = 9
if 1 < x < 10:
    print("Between")

# Equivalent to:
# if x > 1 and x < 10:
````
