Use this notebook to practice Python string fundamentals and real use cases.

## Task 1: Basic

Create string `"Python Programming"`.
Print:
1. First character
2. Last character
3. Length


## Task 2: Slicing

From the above string print:
1. `"Python"`

## Task 3: Methods

Given string `" hello world "`:
1. Remove spaces from both sides

3. Replace `"WORLD"` with `"Python"`


## Task 4: Word Counter

## Task 5: Vowel Counter

Take a string as input and count vowels.

## Task 6: Password Strength

Take password as input.
Rules:
1. If length `< 8` -> `Weak`
2. If no digit -> `Weak`
3. Else -> `Strong`

Use:
1. `len()`
2. `any()`
3. `isdigit()`


## Task 7: Username Validator

2. Must contain only letters, numbers, or underscore

## Task 8: Palindrome Check

## Task 9: Character Frequency

## Task 10: Email Checker

Print `Valid` or `Invalid`.


## Task 11: Censor Bad Words
