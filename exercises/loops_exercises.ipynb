### Task 7 - Real Mini Problem (Password attempts)

- Correct password is `"python"`
- Allow max 3 attempts
- If correct: `"Access granted"`
- Else after 3 attempts: `"Account locked"`
- Use loop + `break`


````python
# Task 7 using for loop
correct_pass = "python"
no_of_attempts = 3
for i in range(no_of_attempts):
    user_attempt = input("Enter password: ")
    if user_attempt == correct_pass:
        print("Access granted")
        break
else:
    print("Account Locked")
````

````python
# Task 7 using while loop
correct_pass = "python"
no_of_attempts = 3
while no_of_attempts != 0:
    user_attempt = input("Enter password: ")
    if user_attempt == correct_pass:
        print("Access granted")
        break
    no_of_attempts -= 1
else:
    print("Account Locked")
````

````python
hours = [2, 3, 1, 4, 2, 5, 3]
total_hours = 0
for i in hours:
   total_hours += i

print(f"Total {total_hours}hr study houre this week")
````

You track daily steps:

steps = [4500, 8000, 7600, 12000, 3000, 9000, 6500]

Write a program that:

Calculates total steps

Finds days above 8000 steps

````python
steps = [4500, 8000, 7600, 12000, 3000, 9000, 6500]

total_steps = sum(steps)

days_above_8000 = 0

for step in steps:
    if step > 8000:
        days_above_8000 += 1

print(f"Total steps: {total_steps}")
print(f"Days above 8000: {days_above_8000}")
````
