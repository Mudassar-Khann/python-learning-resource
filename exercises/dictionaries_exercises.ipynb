### Solved Question


First Non-Repeating Element

Input:

[4,5,1,2,1,2,4]

Output:

5

Use a dictionary to count frequencies.


```python
def non_reapting(num):
    a = {}
    for i in num:
        a[i] = a.get(i, 0) + 1
    for i in a:
        if a[i] == 1:
            return i
    # return a
a = [4,5,1,2,1,2,4]
print(non_reapting(a))
```

Two Sum (Dictionary Version)

Write:

two_sum(nums,target)

Example:

nums = [2,7,11,15]
target = 9

Output:

[0,1]

Constraint:

O(n)

Hint: store numbers in a dictionary.

```python
def two_sum(num: list[int], target: int) -> list[input]:
    s = {x : num.index(x) for x in num if x <= target}
    for i in s:
        temp = target - i
        if temp in s:
            return [s[i], s[temp]]

        else:
            continue




nums = [3,4,2,7,11,15]
print(two_sum(nums,9))
```

Group Words by Length

Input:

["cat","dog","apple","bat","banana"]

Output:

{

3:["cat","dog","bat"],

5:["apple"],

6:["banana"]

}


```python
def group_by(word):
    a = {}
    for i in word:

        temp = len(i)
        # print(i, temp)
        if temp not in a:
            a[temp] = [i]
            continue
            # print(a)
        a[temp].append(i)
    return a




words = ["cat","dog","apple","bat","banana"]


print(group_by(words))
```
