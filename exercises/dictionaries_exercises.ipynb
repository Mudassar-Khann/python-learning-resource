### Solved Question


First Non-Repeating Element

Input:

[4,5,1,2,1,2,4]

Output:

5

Use a dictionary to count frequencies.


```python
def first_non_repeating(nums):

    counts = {}

    for n in nums:
        counts[n] = counts.get(n, 0) + 1

    for n in nums:
        if counts[n] == 1:
            return n
```

Two Sum (Dictionary Version)

Write:

two_sum(nums,target)

Example:

nums = [2,7,11,15]
target = 9

Output:

[0,1]

Constraint:

O(n)

Hint: store numbers in a dictionary.

```python
def two_sum(nums, target):

    seen = {}

    for i, num in enumerate(nums):

        complement = target - num

        if complement in seen:
            return [seen[complement], i]

        seen[num] = i
```

Group Words by Length

Input:

["cat","dog","apple","bat","banana"]

Output:

{

3:["cat","dog","bat"],

5:["apple"],

6:["banana"]

}


```python
def group_by(word):
    a = {}
    for i in word:

        temp = len(i)

        if temp not in a:
            a[temp] = [i]
            continue

        a[temp].append(i)
    return a




words = ["cat","dog","apple","bat","banana"]


print(group_by(words))
```

```python
# cleaner version
def group_by(words):

    groups = {}

    for word in words:

        length = len(word)

        groups.setdefault(length, []).append(word)

    return groups
```
