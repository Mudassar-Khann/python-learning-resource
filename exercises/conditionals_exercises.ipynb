### Task 1 - Even or Odd

- Ask user for a number
- Print `"Even"` or `"Odd"`



```python
# Write your answer here
```

```python
# solution
number = int(input("enter the number"))
if number % 2 == 0:
    print("number is even")
else:
    print("number is odd")
```

```python
# you can also solve it using Using Ternary (One-Line If)

num = int(input("Enter a number: "))

print("Even" if num % 2 == 0 else "Odd")
```

```python
# you can also solve this qustion using Using Bitwise Operator (&)
# cuse the last bit of even numbers are zero while for odd its 1
num = int(input("Enter a number: "))

if num & 1:
    print("Odd")
else:
    print("Even")
```
