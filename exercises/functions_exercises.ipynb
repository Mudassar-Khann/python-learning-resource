````python
# function that returns squre of a number

def square(n):
    return n ** 2

print(square(5))
````

````python
# create function that that checks weather the number is even and return True or false depending on the vale
def is_even(n):
    return n % 2 == 0

print(is_even(3))
````

````python
# write a function that returns the sum of numbers from 1 to n based on the input

def sum_using_loop(n): # Time Complexity O(n)
    total = 0
    for i in range(1 , n+1):
        total += i
    return total
# let's solve it using more efficient method using formula discovered by Carl Friedrich Gauss.
def sum_using_formula(n): # Time Complexity O(1)
    return n * ( n + 1 ) // 2

print(sum_using_loop(5))
print(sum_using_formula(5))
````

````python
# Write a function that return True if number is prime and false if otherwise

def is_prime(n):
    if n < 2:
        return False
    if n > 2 and n % 2 == 0:
        return False
    return (n- n) == 0 or n % 2 != 0


print(is_prime(11))
````

````python
# write a factorial function using loop

def factorial(n):
    fact_result = 1
    for i in range(1,n+1):
        fact_result *= i
    return fact_result


print(factorial(5))
````

````python
# write 2 functions square and cube then write 3rd function that sums the result of both function
def square(n):
    return n ** 2
def cube(n):
    return n ** 3
def sum_of_powers(n):
    return square(n) + cube(n)

print(sum_of_powers(2))
````

````python
# write a function that takes a list and returns count of even number using loop
numbers = [1,2,3,4,5,6]

def even_no(n):
    even_count = 0
    for i in n:
        if i % 2 == 0:
            even_count += 1
    return even_count

print(even_no(numbers))
````

````python
# write a function taht takes a list of study hours and return total_minutes, average_minutes, longest_session
sessions = [30, 45, 60, 20, 90, 50]
def study_stats(n):
    total_minutes = sum(n)
    average_minutes = total_minutes / len(n)
    longeest_session = max(n)

    return total_minutes , round(average_minutes, 2) , longeest_session

print(study_stats(sessions))
````
